```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plot
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
path = r"."
```

```python
dfStartap = pd.read_csv(path + "\Startup.csv")
dfStartap
```

```python
dfStartap.plot.scatter(x='Marketing Spend', y='Profit')
dfStartap.plot.scatter(x='R&D Spend', y='Profit')
dfStartap.plot.scatter(x='Administration', y='Profit')
```

```python
X = dfStartap[['Administration','R&D Spend','Marketing Spend']]
y = dfStartap['Profit']
X_train, X_test, y_train, y_test = train_test_split(X, y)

model = LinearRegression()
model.fit(X_train, y_train)

B0 = model.intercept_
B1, B2, B3 = model.coef_

print(B1, B2, B3)
```
